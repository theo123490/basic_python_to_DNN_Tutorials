##Pandas
Pandas merupakan library pada python yang digunakan untuk pengolahan **Data Frame**. Data frame merupakan struktur data yang mirip dengan tabel dengan kolom dan baris. kolom pada dataframe dapat diberikan nama dan setiap baris merupakan satu index data. 


### Pembuatan Data Frame

 * `markersize` menentukan besarnya mark yang digunakan. 
 * `plt.legend` menentukan legend yang digunakan pada gambar

 * `plt.ylim` menentukan nilai mainimal dan maksimal pada sumbu y
 * `plt.xlim` menentukan nilai mainimal dan maksimal pada sumbu x
 * `plt.ylabel` menentukan nilai mainimal dan maksimal pada sumbu y
 * `plt.xlabel` menentukan nilai mainimal dan maksimal pada sumbu x
 
 
 

```
x = np.arange(0, 2*np.pi, 0.02)
y = np.sin(x)
y1 = np.sin(2*x)
y2 = np.sin(3*x)
ym1 = np.ma.masked_where(y1 > 0.5, y1)
ym2 = np.ma.masked_where(y2 < -0.5, y2)

plt.plot(x, y, linewidth = 4)
plt.plot(x, ym1, '-x', linewidth=2, markevery = 7)
plt.plot( x, ym2, 'o', markersize = 10, markevery = 3)


plt.legend(('No mask', 'Masked if > 0.5', 'Masked if < -0.5'),
           loc='upper right')
plt.title('Masked line demo')

plt.ylim(-1.7,1.2)
plt.xlim(-1,7)
plt.ylabel('ini adalah ylabel')
plt.xlabel('ini adalah xlabel')

plt.show()
```

## Seaborn

Seaborn merupakan salah satu library yang dibentuk bedasarkan library matplotlib. library seaborn mempermudah visualisasi data, terutama data dari suatu dataframe.

### Scatterplot
pada cotoh berikut, dilakukan `scatterplot` yang menunjukan label dengan mudah pada data `iris`. berikut merupakan penjelasan dari fungsi `seaborn.scatterplot` yang digunakan :
 * `data` menentukan dataframe yang digunakan
 * `x` menentukan kolom dataframe yang digunakan sebagai data pada sumbu x
 * `y` menentukan kolom dataframe yang digunakan pada sumbu y
 * `hue` akan membedakan warna dari datapoint pada scatterplot bedasarkan kolom yang dipilih
 * `style` akan membedakan bentuk marker dari datapoint bedasrkan kolom yang dipilih
 * `size` akan membedakan ukuran dari marker bedarakan datapoint yang dipilih
 * `pallete` akan menentukan pallete warna yang digunakan pada gambar 

```
ax = sns.scatterplot(data=iris,
                     x='sepal.length', 
                     y='sepal.width', 
                     hue = 'variety',
                     style = 'variety',
                     size = 'petal.length',
                    #  palette  = 'deep',
                     palette = 'colorblind')
                     
```

Selain itu salah satu kemampuan dari `seaborn` adalah untuk memvisualisasikan beberapa variabel dalam satu gambar. pada contoh berikut digunakan `size` dan `hue` untuk menunjukan data `petal.length` dan `petal.width`. untuk menunjukan warna sebagai data continous maka dibentuk sebuah sistem pewarnaan gradient untuk merepresentasikan suatu nilai. pada seaborn hal ini dapat dicapai menggunakan `seaborn.cubehelix_pallete`.

```
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
sns.scatterplot(x='sepal.length', 
                     y='sepal.width', 
                     hue = 'petal.width',
                     size = 'petal.length',
                     style = 'variety',
                     palette=cmap,
                     data=iris)
```

Pada gambar diatas terlihat bahwa legend yang digunakan terlalu besar pada gambar untuk digunakan sebagai media visualisasi yang baik. Dikarenakan seaborn merupakan library yang berbasis pada library matplotlib, maka fungsi pada matplotlib dapat digunakan untuk  memodifikasi gambar yang digunakan, pada contoh berikut digunakan fungsi `plt.figure(figsize)` untuk menentukan besar dari gambar yang dihasilkan.


```
cmap = sns.cubehelix_palette(dark=0, light=0.8, as_cmap=True)
plt.figure(figsize=(15, 10))
sns.scatterplot(x='sepal.length', 
                     y='sepal.width', 
                     hue = 'petal.width',
                     size = 'petal.length',
                     style = 'variety',
                     palette=cmap,
                     data=iris)
```

## regplot

untuk menentukan korelasi antara data `regplot` dapat digunakan untuk menunjukan korelasi antara data. Selain itu regplot dapat menunjukan nilai confidence level dari nilai regressi yang ada. `regplot` sangat berguna untuk melihat hubungan data pada secara singkat.

Pada data pembelajaran `regplot` digunakan data `tips` yang sudah disiapkan oleh seaborn

```
tips = sns.load_dataset("tips")
tips
```

Berikut merupakan contoh regplot dari data `total_bill` dengan harga `tip` dengan warna hijau

```
plt.figure(figsize=(10, 5))
sns.regplot(x="total_bill", y="tip", data=tips, color = 'g')
```

Berikut merupakan contoh regplot dari data `Size` dengan harga `tip` dengan warna hijau

```
plt.figure(figsize=(10, 5))
sns.regplot(x="size", y="tip", data=tips)
```

Terlihat bahwa data `size` bertumpuk dan menyebabkan pembacaan data menjadi lebih sulit, salah satu metode untuk mempermudah visualisasi penyebaran data adalah mengunakan `x_estimator` seperti berikut

```
plt.figure(figsize=(10, 5))
sns.regplot(x="size", y="tip", data=tips, x_estimator=np.mean)
```

data yang tidak bertumpuk seperti pada data `total_bill` juga dapat disimplifikasi menggunakan `x_bins` untuk mengumpulkan data-data yang berdekatan 

```
plt.figure(figsize=(10, 5))
sns.regplot(x="total_bill", y="tip", data=tips, color = 'g', x_bins=10)
```

## Bar Plot

```
titanic = sns.load_dataset("titanic")
titanic
```

```

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=crashes,
            label="Total", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="alcohol", y="abbrev", data=crashes,
            label="Alcohol-involved", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",
       xlabel="Automobile collisions per billion miles")
sns.despine(left=True, bottom=True)
```
