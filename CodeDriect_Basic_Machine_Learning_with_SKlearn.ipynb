# Principal Component Analysis (PCA)

Principal componen analysis merupakan salah satu metode reduksi dimensi yang umum digunakan. berikut merupakan gambaran dari bagaimana cara PCA dilakukan

![alt text](https://miro.medium.com/max/602/1*4HxMtrguXVYqCnQP711nxQ.png)

![alt text](https://miro.medium.com/max/394/1*wSzOwnF_Rk1Bhup5ZAJROg.png)


pada contoh berikut digunakan dataset data spektrum reflektansi dari beberapa kulit manusia. kolom `subject` menunjukan manusia yang berbeda-beda

```
skin = pd.read_csv('https://raw.githubusercontent.com/theo123490/Human_Skin_Spectrum/master/SkinReflectance_pca.csv')
skin
```

terlihat bahwa data pada kolom pertama merupakan data `Unnamed: 0` yang menandakan bahwa kolom pertama meruapakan kolom yang seharusnya digunakan sebagai data index. maka pada pembacaan csv dapat digunakan `index_col`

```
skin = pd.read_csv('https://raw.githubusercontent.com/theo123490/Human_Skin_Spectrum/master/SkinReflectance_pca.csv',index_col=0)
skin
```

menunjukan beberapa grafik spectrum dari berbagai spektrum dari manusia yang berbeda. terlihat bahwa spektrum sangatlah mirip, maka pembedaan sangat sulit untuk dilakukan secara manual

```
plt.plot(skin['2'].drop('subject').values, color = 'r')
plt.plot(skin['5'].drop('subject').values, color = 'g')
plt.plot(skin['13'].drop('subject').values, color = 'b')
plt.plot(skin['18'].drop('subject').values, color = 'k')
```

untuk melakukan principal component analysis (PCA) setiap parameter harus diletakan pada kolom, maka untuk menyesuaikan data dari `skin` dilakukan transpose

```
skin = skin.T
skin
```

untuk mengolah data dalam bentuk PCA, maka data dengan label harus dihapuskan untuk melakukan fitting pada PCA

```
x = skin.drop(columns = 'subject')
```

berikut merupakan metode PCA dengan 6 principal component

```
from sklearn.decomposition import PCA
pca = PCA(n_components=6)
principalComponents = pca.fit_transform(x)


principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6']
             , index = x.index)
pca_df = pd.concat([principalDf, skin[['subject']]], axis = 1)
pca_df
```

untuk dengan mudah mengevaluasi PCA pada seluruh Principal component maka dilakukan `pairplot`. Terlihat bahwa data `pc1` dan `pc2` memilikin perbedaan yang sangat jelas yang menunjukan perbedaan yang baik pada setiap kelas. pada distribution plot terlihat bahwa variabilitas data lebih cepat menumpuk antara satu sama lain. hal ini sesuai dengan teori PCA dimana data principal component awal semakin penting dibandingkan data principal component akhir. namun semakin dikit data principal component sedikit informasi yang didapatkan

```
sns.pairplot(pca_df, hue='subject')
```

Berikut merupakan scatterplot yang lebih jelas untuk `pc1` dan `pc2`

```
sns.scatterplot(data = pca_df, x = 'pc1', y='pc2',style='subject',hue='subject')
```

# Classification

Classification merupakan salah satu tujuan machine learning untuk mengklasifikasi suatu data pada suatu kelas. pada klasifikasi ini akan digunakan dataset `iris` yang telah disiapkan oleh `seaborn`

```
iris = sns.load_dataset('iris')
iris.head()
```

Sebelum melakukan klasifikasi, salah satu tahap yang baik untuk dilakukan adalah melakukan visualisasi pada data yang dimiliki, untuk data dengan input yang tidak banyak, `pairplot` umum dilakukan untuk memahami data yang ada. 

untuk latihan klasifikasi, agar lebih mudah untuk dipahami, klasifikasi dilakukan pada 2 parameter saja. terlihat data `petal_length` dan `petal_width` menunjukan perubahan data yang cukup baik. **pada umumnya, data `petal_length` dan `petal_width` akan digunakan untuk klasifikasi karena distinksi data yang baik**. namun, untuk **keperluan latihan** ini, digunakan data `sepal_width` dan `sepal_length` agar terdapat overlapping yang dapat digunakan untuk diskusi lebih lanjut

```
sns.pairplot(iris,hue='species')
```

### Splitting data

pada machine learning `x` umum digunakan sebagai input dan `y` umum digunakan sebagai target output.

```
# x = iris[['petal_length','petal_width']]
x = iris[['sepal_length','sepal_width']]
y = iris['species']
```

pada machine learning data yang dimiliki dipisahkan menjadi data **TRAIN** dan data **TEST**. Data train digunakan untuk melatih model ang digunakan oleh mesin. data test digunakan untuk menguji model yang digunakan. hal ini dilakukan untuk mencegah terjadinya **overfitting** pada model. pada umumnya data **train** dan **test** dipisahkan dalam ration  train:test sebesar **3:1 (75:35) atau 2:1 (67:33)**

![alt text](https://miro.medium.com/max/2400/1*UCd6KrmBxpzUpWt3bnoKEA.png)

**overfitting** merupakan fenomena dimana model menyesuaikan data terhadap model terlalu jauh yang menyebabkan kesalahan lebih tinggi pada daset lainya.

Selain memecah data menjadi data **Train** dan **Test**, untuk pengembangan data lebih lanjut, dibentuk data **development** yang digunakan untuk menguji **model akhir** untuk mencegah overfitting dan mengetahui akurasi model lebih baik

pada `sklearn`, memecahkan data menjadi train dan test dapat dengan mudah dilakukan menggunakan **`sklearn.model_selection.train_test_split`**

`test_size` menunjukan persentase data yang ingin diubah menjadi data test

`random_state` merupakan nilai acak yang dipilih untuk memisahkan data pada `train_test_split`

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=32)
```

## K-Nearest Neighbor
K-nearest neighbor atau lebih umum disebut KNN merupakan metode machine learning yang membandingkan suatu datapoint pada datapoint lainya bedasarkan kedekatan data. terdapat beberapa metode KNN, dua metode yang umum digunakan merupakan metode KNN bedasarkan **Centroid** dan **Voting**

### Nearest Centroid

![alt text](https://www.mathworks.com/help/examples/stats/win64/PartitionDataIntoTwoClustersExample_02.png)

Centroid merupakan pusat dari cluster data. centroid merepresentasikan probabilitas kelas data bedasarkan lokasi dari suatu data. semakin dekat data terhadap satu centroid maka data dianggap sebagai kelas tersebut

untuk melakukan KNN nearest centroid, dapat digunakan `sklearn.neighbors.NearestCentroid`

```
from sklearn.neighbors import NearestCentroid

model = NearestCentroid()
model.fit(x_train, y_train)
print(model.centroids_)
```

terdapat beberapa parameter yang umum digunakan untuk menunjukan tingkat keberhasilan dari suatu klasifikasi data. parameter yang umum digunakan merupakan nilai:


*   Akurasi
*   Precision
*   Recall
*   F1-score

**Akurasi** : akurasi menunjukan persentase data berhasil diklasifikasikan dengan benar

**Precision** : Precision menunjukan presisi data dalam menentukan kelas suatu data. precision yang rendah dapat mengindikasikan model memilih banyak atau seluruh data sebagai nilai positif. Precision dapat dihitung menggunakan persamaan berikut :
\begin{equation*}
\frac{TP}{FP+TP}
\end{equation*}

**Recall** : Recall menunjukan nilai banyaknya data positif yang berhasil dideteksi oleh model. recall yang rendah menandakan model tidak sukses dalam mendeteksi kelas tersebut. recall dapat dihitung menggunakan persamaan berikut.

\begin{equation*}
\frac{TP}{FN+TP}
\end{equation*}

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/2/26/Precisionrecall.svg/350px-Precisionrecall.svg.png)


**F1-score** : f1-score merupakan parameter yang menghitung keseimbangan nilai precision dan recall dengan persamaan berikut:


alt text


Selain itu, pada sistem klasifikasi multi-class parameter yang didapatkan juga dapat dihitung sebgai **macro average** dan **weighted average**. **macro average** akan merata-ratakan nilai yang didaptkan setelah seluruh nilai sudah dihitung. sedangkan weighted average menghitung rata-rata parameter bedasarkan banyaknya data yang dimiliki dan memberikan **beban** yang sesuai pada setiap parameter



---

pada `sklearn` evaluasi model dapat dengan mudah dilakukan menggunakan `sklearn.metrics.classification_report`

nilai `support` menunjukan banyaknya data setiap kelas yang digunakan pada data test yang dimiliki

```
from sklearn.metrics import classification_report
predictions = model.predict(x_test)

print(classification_report(y_test, predictions, 
                            target_names=y.unique().astype(str)))
```

untuk memprediksi data dengan model yang sudah ada, dapat digunakan `model.predict` menggunakan data yang dimiliki

```
from sklearn.metrics import classification_report
predictions = model.predict([[4.7 , 2.12],[7.2,4.2]])
print(predictions)
```

metode centroid tidak bekerja lebih baik untuk seluruh data, pada data lingkaran berikut metode centroid tidak akan berhasil dikarenakan lokasi centroid akan berada pada lokasi yang hampir sama


```
from sklearn.datasets import make_circles
circular_x, circular_y = make_circles(n_samples=400, factor=.3, noise=.05)

sns.scatterplot(circular_x[:,0],circular_x[:,1],hue=circular_y)
plt.show()

circular_x_train, circular_x_test, circular_y_train, circular_y_test = train_test_split(circular_x, circular_y, test_size=0.33, random_state=32)

circular_model = NearestCentroid()
circular_model.fit(circular_x_train, circular_y_train)
print(circular_model.centroids_)

circular_predictions = circular_model.predict(circular_x_test)

print(classification_report(circular_y_test, circular_predictions))

sns.scatterplot(circular_x_test[:,0],circular_x_test[:,1],hue=circular_predictions)
plt.scatter(circular_model.centroids_[0],circular_model.centroids_[1],marker = 'x')
plt.show()
```

### KNN voting

Metode voting pada KNN dilakukan dengan melakukan voting bedasarkan tetangga yang berdekatan. setiap kelas akan memberikan 1 

![alt text](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1531424125/KNN_final1_ibdm8a.png)

```
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=6)
model.fit(x_train, y_train)
```

```
from sklearn.metrics import classification_report
predictions = model.predict(x_test)

print(classification_report(y_test, predictions,
    target_names=y.unique().astype(str)))
```

```
circular_x, circular_y = make_circles(n_samples=400, factor=.3, noise=.05)

sns.scatterplot(circular_x[:,0],circular_x[:,1],hue=circular_y)
plt.show()

circular_x_train, circular_x_test, circular_y_train, circular_y_test = train_test_split(circular_x, circular_y, test_size=0.33, random_state=32)

circular_model = KNeighborsClassifier(n_neighbors=6)
circular_model.fit(circular_x_train, circular_y_train)

circular_predictions = circular_model.predict(circular_x_test)

print(classification_report(circular_y_test, circular_predictions))

sns.scatterplot(circular_x_test[:,0],circular_x_test[:,1],hue=circular_predictions)
plt.show()
```

## Support Vector Machine (SVM)

![alt text](https://miro.medium.com/max/1200/1*06GSco3ItM3gwW2scY6Tmg.png)
Salah satu metode klasifikasi lain adalah menggunakan SVM (Support Vector  Machine). SVM membuat mencari hyperplane atau pembatas pada dataset yang akan memisahkan kelas-kelas dari suatu data. nilai optimal hyperplane didapatkan dari besarnya margin yang didapatkan

```
from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)
```

```
from sklearn.metrics import classification_report
predictions = model.predict(x_test)

print(classification_report(y_test, predictions,
    target_names=y.unique().astype(str)))
```

Menggunakan kernel trick, SVM juga dapat mengolah data circular. pada data circular berikut, digunakan persamaan 

\begin{equation*}
z = y^2 +x^2
\end{equation*}

yang menghasilkan data seperti berikut:
![alt text](https://miro.medium.com/max/2800/0*ngkO1BblQXnOTcmr.png)

```
circular_x, circular_y = make_circles(n_samples=400, factor=.3, noise=.05)

sns.scatterplot(circular_x[:,0],circular_x[:,1],hue=circular_y)
plt.show()

circular_x_train, circular_x_test, circular_y_train, circular_y_test = train_test_split(circular_x, circular_y, test_size=0.33, random_state=32)

circular_model = SVC()
circular_model.fit(circular_x_train, circular_y_train)

circular_predictions = circular_model.predict(circular_x_test)

print(classification_report(circular_y_test, circular_predictions))

sns.scatterplot(circular_x_test[:,0],circular_x_test[:,1],hue=circular_predictions)
plt.show()
```

#Clustering

Clustering merupakan salah satu metode unsupervised learning yang digunakan untuk memisah data yang dimiliki menjadi berbagai kelas. pada latihan ini akan dijelaskan clustering bedasarkan metode KNN. clustering dilakukan dengan menggerakan centroid hingga didapatkan pembagian kelas yang sesuai. 

terlihat bila dibandingkan dengan data sesungguhnya beberapa bagian cluster tidak sesuai dengan kelas sesungguhnya. hal ini disebabkan oleh adanya daerah cluster yang bertumpukan. pada kasus clustering kelas dibentuk dengan boundary line yang jelas yang menyebabkan data overlapping tidak dapat dipisahkan


```
from sklearn.cluster import KMeans
```
