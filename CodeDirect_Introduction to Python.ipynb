`object` dapat digunakan untuk membuat suatu item pada code yang ingin dimanipulasikan seperti object `Circle` dan `Rectangle` berikut:

````
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class Circle(object):
  '''Ini merupakan suatu object lingkaran'''
  def __init__(self,radius,color):
    self.radius = radius
    self.color = color

  def add_radius(self,r):
    self.radius = self.radius+r

  def draw_circle(self):

    ax = plt.figure(figsize=(5,5))

    circle1 = plt.Circle((0,0), self.radius, color=self.color)
    ax = plt.gca()

    ax.cla() # clear things for fresh plot

    # change default range so that new circles will work
    ax.set_xlim((-(self.radius+5), (self.radius+5)))
    ax.set_ylim((-(self.radius+5), (self.radius+5)))

    ax.add_artist(circle1)
  
  def __repr__(self):
    return 'lingkaran dengan radius: ' +str(self.radius)+' dengan warna: ' +str(self.color)
````

````

class Rectangle (object):
    def __init__(self,color,height,width):
        self.height = height
        self.width = width
        self.color = color
    def draw_rect(self):

        ax = plt.figure(figsize=(5,5))
        
        rect = patches.Rectangle((0,0), self.width,self.height, color=self.color)
        ax = plt.gca()

        ax.cla() # clear things for fresh plot
        # change default range so that new circles will work
        ax.set_xlim((-4, (self.width*1.3)))
        ax.set_ylim((-4, (self.height*1.3)))
        
        ax.add_patch(rect)
````

Terlihat terdapat fungsi `__init__`:
fungsi `__init__` menunjukan fungsi utama dari object dimana saat object dibentuk fungsi `__init__` dijalankan terlebih dahulu. fungsi `__init__` umum digunakan untuk menentapkan parameter dari object seperti parameter `radius` dan `color` pada object `Circle`
